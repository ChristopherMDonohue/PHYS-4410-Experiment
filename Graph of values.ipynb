```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
methods=np.array(['Final Deflection','Curve Fit','Initial Acceleration'])
values=np.array([6.29E-11,6.69E-11,7.27E-11])
uncer=np.array([5.80E-12,5.70E-12,7.46E-12])
```

```python
plt.scatter(methods,values,label='Experimentally Measured Values')
G=np.array([6.67430E-11,6.67430E-11,6.67430E-11])
plt.plot(methods,G,color='red',label='Currently Accepted Value')
plt.errorbar(methods,values,yerr=uncer,linestyle='None')
plt.fill_between(methods,G,G+.00015E-11,color='yellow')
plt.title('Values and Uncertainty for Each Method')
plt.xlabel('Methods')
plt.ylabel('Measured value of G ($kg^-1m^3s^-2$)')
plt.legend(loc="upper left")
```
